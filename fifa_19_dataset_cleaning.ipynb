**DELETE 'LOANED FROM',CONTAIN 93% MISSING VALUE**

```python
df=df.drop(['Loaned From'],axis=1)
```

**CHECKING DATATYPE**

```python
df.dtypes[df.isnull().any()]
```

***FILLING THE MISSING VALUE ***

```python
df['Jersey Number'].fillna('NA')
```

```python
list1=['Club', 'Position', 'Joined', 'Contract Valid Until', 'Release Clause']
for i in list1:
  df[i] =df[i].fillna(df[i].mode()[0])
```

```python
list2='LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB'
for i in list2:
  df[i] = pd.to_numeric(df[i], errors='coerce').fillna(0).astype(int)
```
